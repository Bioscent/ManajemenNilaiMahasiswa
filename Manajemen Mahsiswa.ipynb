```python
data_mahasiswa = {}

def data_nilai_akhir(tugas, uts, uas):
    return tugas * 0.2 + uts * 0.4 + uas * 0.4

def tentukan_grade(nilai_akhir):
    if nilai_akhir >= 85:
        return 'A'
    elif nilai_akhir >= 70:
        return 'B'
    elif nilai_akhir >= 55:
        return 'C'
    elif nilai_akhir >= 40:
        return 'D'
    else:
        return 'E'

def input_data_mahasiswa():
    print("\n========== INPUT NILAI ==========")
    matkul = input("Masukkan nama mata kuliah: ")

    if matkul not in data_mahasiswa:
        data_mahasiswa[matkul] = []

    nama = input("Masukkan nama mahasiswa: ")
    tugas = float(input("Masukkan nilai tugas: "))
    uts = float(input("Masukkan nilai UTS: "))
    uas = float(input("Masukkan nilai UAS: "))

    nilai_akhir = data_nilai_akhir(tugas, uts, uas)
    grade = tentukan_grade(nilai_akhir)

    data_mahasiswa[matkul].append({
        'nama': nama,
        'tugas': tugas,
        'uts': uts,
        'uas': uas,
        'nilai_akhir': nilai_akhir,
        'grade': grade
    })

    print(f"Data mahasiswa {nama} untuk mata kuliah {matkul} berhasil ditambahkan.\n")

def tampilan_tabel(daftar):
    if not daftar:
        print("Tidak ada data.\n")
        return

    print("+----+----------------------+--------+-------+-------+-------------+--------+")
    print("| No | Nama                 | Tugas  | UTS   | UAS   | Nilai Akhir | Grade  |")
    print("+----+----------------------+--------+-------+-------+-------------+--------+")

    for i, m in enumerate(daftar, 1):
        print(
            f"| {i:<2} "
            f"| {m['nama']:<20} "
            f"| {m['tugas']:<6} "
            f"| {m['uts']:<5} "
            f"| {m['uas']:<5} "
            f"| {m['nilai_akhir']:<11.2f} "
            f"| {m['grade']:<6} |"
        )

    print("+----+----------------------+--------+-------+-------+-------------+--------+")

def tampilkan_semua_data():
    print("\n========== SEMUA DATA NILAI ==========")

    if not data_mahasiswa:
        print("Belum ada data sama sekali.\n")
        return

    for matkul, daftar in data_mahasiswa.items():
        print(f"\n--- Mata Kuliah: {matkul} ---")
        tampilan_tabel(daftar)

def tampilkan_per_matkul():
    print("\n========== LIHAT PER MATA KULIAH ==========")
    matkul = input("Masukkan nama mata kuliah: ")

    if matkul not in data_mahasiswa:
        print("Data mata kuliah tidak ditemukan.\n")
        return

    print(f"\n--- Daftar Nilai Mata Kuliah: {matkul} ---")
    tampilan_tabel(data_mahasiswa[matkul])

def main():
    while True:
        print("""
===== MENU =====
1. Input Data Mahasiswa
2. Tampilkan Semua Daftar Nilai
3. Tampilkan Nilai per Mata Kuliah
4. Keluar
""")

        pilihan = input("Pilih menu: ")

        if pilihan == '1':
            input_data_mahasiswa()
        elif pilihan == '2':
            tampilkan_semua_data()
        elif pilihan == '3':
            tampilkan_per_matkul()
        elif pilihan == '4':
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid.\n")

main()
```
